### Add Color, Marker and Line to the plot

If you look at the arguments plt.plot() expects, you will see it expects 3 basic arguments in the following order: (x, y, format).

You are well clear about x and y. Let's look at the last argument format. This format is a short hand combination of {color}{marker}{line}.
The first placeholder in the format will get colour, followed by the marker followed by the line style.

Let's understand it with help of an example. 

Let's say you wish to a line plot with dots and the colour should be green. You need to specify the format as 'go-'. The format 'go-' has 3 characters standing for: ‘green colored dots with solid line’.  

Let's draw it

````python
# 'go-' stands for ‘green colored dots with solid line’.
plt.plot([1,2,3,4,5], [1,2,3,4,10], color="#0A0")
plt.show()
````

Now let’s draw a scaterplot with greendots.

By omitting the line part (‘-‘) in the end, you will be left with only green dots (‘go’), which makes it draw a scatterplot.

````python
# 'go' stands for green dots
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go')
plt.show()
````

Explicit the arguments

````python
# More explicitly:
plt.plot([1,2,3,4,5], [1,2,3,4,10], color="g", marker="o", linestyle="-", markersize=10)
plt.show()
````

Few commonly used short hand format examples are:
* 'r*--' : ‘red stars with dashed lines’
* 'ks.' : ‘black squares with dotted line’ (‘k’ stands for black)
* 'bD-.' : ‘blue diamonds with dash-dot line’.

For a complete list of colors, markers and linestyles, check out the help(plt.plot) command.

### Modify the Axis Ticks Positions and Labels

There are 2 basic things you will probably ever need in matplotlib when it comes to manipulating axis ticks:
1. How to control the position and tick labels? (using plt.xticks())
2. How to control which axis’s ticks (top/bottom/left/right) should be displayed (using plt.tick_params())

````python
from matplotlib.ticker import FuncFormatter

plt.figure(figsize=(12,7), dpi=100) # to get a large figure

plt.plot(area_loc1, price_loc1, 'go', label = 'Locality A')
plt.plot(area_loc2, price_loc2, 'b*', label = 'Locality B')

plt.title('Price vs Area Scatterplot', fontsize=20)
plt.xlabel('Area of the house', fontsize=12)
plt.ylabel('Price of the house', fontsize=12)
plt.legend(loc='best')

# 1. Adjust x axis Ticks
plt.xticks(ticks=np.arange(0, 500, 50), fontsize=12, rotation=30, ha='center', va='top')  # 1 radian = 57.2985 degrees
plt.yticks(ticks=np.arange(0, 6000, 500))

# 2. Tick Parameters
plt.tick_params(axis='both', bottom=True, top=True, left=True, right=True, 
                direction='in', which='major', grid_color='blue')

plt.grid(linestyle='--', linewidth=0.5, alpha=0.55)
plt.show()
````

As you can see I have used some additional commands as well. Let's see what are these used for:

- plt.figure(figsize=(12,7), dpi=100) - to specify the figure size and dpi of the plot
- fontsize - to decide the font size of the texts

- rotation - to rotate the ticks on the axis or some other text in the plot
- plt.grid() - to enable the grids inside the plot, 
- linewidth - to decide the width of lines (grid lines in this example)
- alpha - to decide on the plot fill opacity (from 0 to 1)
- xticks - Values to use for x-axis ticks
- tick_params - Change the appearance of ticks, tick labels, and gridlines.

## Draw subplots 

Various plots in the matplotlib resides inside a figure object. 

![Plot](https://www.machinelearningplus.com/wp-content/uploads/2019/01/99_matplotlib_structure-1-640x646.png)

You already know how to create a new figure. i.e using `plt.figure`. You can create multiple subplots in a figure. Or I should say you can create a grid of subplots in a figure.

There are 2 ways to create subplots:
1. You define the grid in the start itself. Example :  `fig, axes = plt.subplots(2, 3)`. It will create a grid of 6 suplots and these can be accessed using axes[i,j] i and j are the index of the subplots. The catch here is that it has created all the placeholders in the start itself. So if you don't use any of the subplots, there would be an empty box at that place.  
2. You create the empty figure first and then keep on adding plots one by one. Example : `fig = plt.figure(figsize=(10,4), dpi=120)` It will create an empty figure. Till now it doesn't know what's coming to the plots.  `ax1 = fig.add_subplot(2, 2, 1)` this command will create the first plot in the figure of 2*2 plots

Quite confusing right?

Let's see it with an example:

````python
area_loc1
````

````python
price_loc1
````

````python
# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,4), dpi=120)

# Plot
axes[0,0].plot(area_loc1, price_loc1, 'go')
axes[0,1].plot(area_loc2, price_loc2, 'b*')
axes[1,0].plot(area_loc2, price_loc2, 'b*')
````

Let's look at the 2nd way of creating subplots

````python
fig = plt.figure(figsize=(10,4), dpi=120)
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

# Plot
ax1.plot(area_loc1, price_loc1, 'go')
ax2.plot(area_loc2, price_loc2, 'b*')
ax3.plot(area_loc2, price_loc2, 'b*')
````

I hope it's clear now. 

Other commands like y_label, title etc also needs a slight modification now. 

The ax1, ax2 and ax3 objects, like plt, has equivalent set_title, set_xlabel and set_ylabel functions. Infact, the plt.title() actually calls the current axes set_title() to do the job.

- plt.xlabel() - ax.set_xlabel()
- plt.ylabel() - ax.set_ylabel()
- plt.title() - ax.set_title()

Now let's jump to our previous example of price vs area in locality

Suppose, You want to draw our two sets of points (locality A and locality B) in two separate plots side-by-side instead of the same plot. How would you do that?

You can do that by creating two separate subplots, aka, axes using `plt.subplots(1, 2)`. This creates and returns two objects:

1. Figure
2. Axes


Previously, I called `plt.plot()` to draw the points. Since there was only one axes by default, it drew the points on that axes itself.

But now, since you want the points drawn on different subplots (axes), you have to call the plot function in the respective axes (ax1 and ax2 in below code) instead of plt.

Notice in below code, I call `ax1.plot()` and `ax2.plot()` instead of calling `plt.plot()` twice.

````python
# Create Figure and Subplots
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4), sharey=False, dpi=120)

# Plot
ax1.plot(area_loc1, price_loc1, 'go')
ax2.plot(area_loc2, price_loc2, 'b*')

# Title, X and Y labels, X and Y Lim
ax1.set_title('Price vs Area Scatterplot Locality A'); ax2.set_title('Price vs Area Scatterplot Locality B')
ax1.set_xlabel('Area of the house');  ax2.set_xlabel('Area of the house')  # x label
ax1.set_ylabel('Price of the house');  ax2.set_ylabel('Price of the house')  # y label
ax1.set_ylim(2000, 4000)
ax2.set_ylim(0, 6000)

fig.suptitle("Overall Title")
fig.tight_layout()

plt.show()
````

Setting sharey=True in plt.subplots() shares the Y axis between the two subplots.

And dpi=120 increased the number of dots per inch of the plot to make it look more sharp and clear. You will notice a distinct improvement in clarity on increasing the dpi especially in jupyter notebooks.

## Add annotations to the plot

You can add texts and annotations to the plot using `plt.annotate` function

````python
# Create Figure and add Annotations
plt.figure(figsize=(12,7))
plt.plot(area_loc1, price_loc1, 'go', label = 'Locality A')
plt.plot(area_loc2, price_loc2, 'b*', label = 'Locality B')
plt.title('Price vs Area Scatterplot', fontsize=20)
plt.xlabel('Area of the house', fontsize=12)
plt.ylabel('Price of the house', fontsize=12)
plt.legend(loc='best')
# 
# Annotation
plt.annotate('Price B < Price A', xy =(150, 1300 + 100),
                xytext =(150, 1300 + 500), 
                arrowprops =dict(facecolor='black', headwidth=5, width=2, headlength=4),
                horizontalalignment='left', verticalalignment='top')


plt.show()
````

__Annotate the points__

````python
# Create Figure and add Annotations
plt.figure(figsize=(12,7))
plt.plot(area_loc1, price_loc1, 'go', label = 'Locality A')
plt.plot(area_loc2, price_loc2, 'b*', label = 'Locality B')
plt.title('Price vs Area Scatterplot', fontsize=20)
plt.xlabel('Area of the house', fontsize=12)
plt.ylabel('Price of the house', fontsize=12)
plt.legend(loc='best')

# Annotate all points
for i, val in enumerate(price_loc2):
    plt.annotate(str(val), xy =(area_loc1[i], val),
                xytext =(area_loc1[i], val + 150), 
                horizontalalignment='center', 
                verticalalignment='center')

plt.show()
````

````python
# Import Data
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/economics.csv")
df.head()
````

Date is the x variable and psavert is the y variable. Let's store these in variables x and y

Now let's draw the line plot

````python
# Draw Plot
plt.figure(figsize=(16,10), dpi= 80)
plt.plot(x, y, color='tab:red')

# Decoration
plt.title('Personal Savings Rate', fontsize=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Personal Savings Rate', fontsize=12)
plt.xticks(np.arange(0, len(x), 60))

plt.show()
````

__Dual Axis Line Plot__

When you want to analyze the effect of one variable on another over time, use dual axis line plot

Task:
- To analyze the effect of unemployement rate on the personal savings rate over time

````python
# Store the values in variables x, y1 and y2
x = df['date']; y1 = df['psavert']; y2 = df['unemploy']
````

Now let's draw the line plot

````python
# Plot Line1 (Left Y Axis)
fig, ax1 = plt.subplots(1,1,figsize=(16,9), dpi= 80)
ax1.plot(x, y1, color='tab:red')

# Plot Line2 (Right Y Axis)
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(x, y2, color='tab:blue')

# Decorations
# ax1 (left Y axis)
ax1.set_xlabel('Year', fontsize=20)
ax1.tick_params(axis='x', rotation=0, labelsize=12)
ax1.set_ylabel('Personal Savings Rate', color='tab:red', fontsize=20)
ax1.tick_params(axis='y', rotation=0, labelcolor='tab:red' )
ax1.grid(alpha=.4)

# ax2 (right Y axis)
ax2.set_ylabel("# Unemployed (1000's)", color='tab:blue', fontsize=20)
ax2.tick_params(axis='y', labelcolor='tab:blue')
ax2.set_xticks(np.arange(0, len(x), 60))
# ax2.set_xticklabels(x[::60], rotation=90, fontdict={'fontsize':10})
ax2.set_title("Personal Savings Rate vs Unemployed: Plotting in Secondary Y Axis", fontsize=22)
fig.tight_layout()
plt.show()
````

Bar chart is a classic way of visualizing items based on counts or any given metric. Or it can be used to view the composition

It represents categorical data with the category on the x axis and values on the y axis (height of the bars) and vice-versa (in case of horizontal bar plot). The bars can be plotted vertically or horizontally. A bar graph shows comparisons among discrete categories.

__When to use bar plot?__

Bar plots are used to:
- Analyze the composition
- Represent the categorical data based on the counts or values
- Compare the discrete categories  

__Various types of bar plot__
- Vertical
- Horizontal
- Multiple Bars by category
- Stacked

__Use Cases__
- Compare various manufacturing companies based on the number of unique vehicles they manufacture 
- Compare the sales of various products

__Vertical Bar Plot__

In case of vertical bar plots, the vertical axis represents the count/values and the horizontal axis represents the categories

Task:
- To compare the number of unique vehicles manufactured by various automobile companies

````python
import pandas as pd
````

````python
# Import Data
df_raw = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")
df_raw.head()
````

Let's get the count of vehicles manufactured by various companies

````python
# Prepare Data
df = df_raw.groupby('manufacturer').size().reset_index(name='counts')
df.head()
````

Now let's draw the bar plot

````python
# Plot Bars
plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df['manufacturer'], df['counts'], width=.5)

# Add the count on top of bar plot
for i, val in enumerate(df['counts'].values):
    plt.text(i, val, float(val), 
             horizontalalignment='center', verticalalignment='bottom', 
             fontdict={'fontweight':500, 'size':12})

# Decoration
plt.gca().set_xticklabels(df['manufacturer'], rotation=60, horizontalalignment= 'right')
plt.title("Number of Vehicles by Manaufacturers", fontsize=22)
plt.xlabel('Manufacturer', fontsize=12)
plt.ylabel('# Vehicles', fontsize=12)
plt.ylim(0, 45)
plt.show()
````

__Building the same in Pandas__

````python
# Vertical Bar Plot in Pandas
ax = df.plot(x ='manufacturer', y='counts', kind = 'bar', 
             figsize=(10,7), width=0.4, 
             xlabel="Manufacturer", 
             ylabel="Count", 
             title="Sales Summary")

for i, val in enumerate(df['counts'].values):
    ax.text(i, val, float(val), 
            horizontalalignment='center', 
            verticalalignment='bottom', 
            fontdict={'fontweight':500, 'size':8})
````

__Horizontal Bar Chart__

It is similar to vertical bar chart. But it's represented horizontally and ordered by the x axis (in most of the cases) 

In case of vertical bar plots, the vertical axis represents the count/values and the horizontal axis represents the categories

All you have to do is use `plt.barh` insted of `plt.bar`

````python
# Prepare Data
df = df_raw.groupby('manufacturer').size().reset_index(name='counts')
df = df.sort_values('counts', ascending=True)
df.head()
````

````python
# Plot Bars
plt.figure(figsize=(16,10), dpi= 80)
plt.barh(df['manufacturer'], df['counts'])

# Add the count on top of bar plot
for val, i in enumerate(df['counts'].sort_index(ascending=False).sort_values(ascending=True).values):
    plt.text(i, val, float(i), 
             horizontalalignment='left', verticalalignment='center', 
             fontdict={'fontweight':500, 'size':12})

# Decoration
plt.gca().set_yticklabels(df['manufacturer'], rotation=60, horizontalalignment= 'right')
plt.title("Number of Vehicles by Manaufacturers", fontsize=22)
plt.xlabel('Manufacturer', fontsize=12)
plt.ylabel('# Vehicles', fontsize=12)
plt.xlim(0, 45)
plt.show()
````

__Multiple Bars by Category - Grouped Bar Plot__

Grouped bar chart or multi-series bar chart or clustered bar chart is an extention to the bar chart. It plots the values for the levels of two categorical variables instead of one. 

The plotting function is the same, all you have to do is transform the data to required format

__Coloured Scatter Plot for various categories__

Use `plt.scatter()` inside a for loop iterating over all categories and use `c =` argument to specify colour for each category 

Task:
- Check the population spread across area in midwest area, colour the dots with category

or in other words

- Find the relation between population and area and colour the dots with category

````python
# Create as many colors as there are unique midwest['category']
categories = np.unique(midwest['category'])
colors = [plt.cm.tab10(i/float(len(categories)-1)) for i in range(len(categories))]
````

````python
colors
````

Let's draw the plot

````python
# Draw Plot for Each Category
fig, axes = plt.subplots(1,1, figsize=(10,7), dpi=120)

for i, category in enumerate(categories):
    midwest.loc[midwest.category==category, :].plot(x='area', 
            y='poptotal', 
            kind='scatter', 
            ax=axes,
            s=20, 
            color=colors[i], 
            label=str(category))
    
    
# Decorations based on axes
plt.gca().set(xlim=(0.0, 0.1), ylim=(0, 90000),
              xlabel='Area', ylabel='Population')

plt.xticks(fontsize=8); plt.yticks(fontsize=8)
plt.title("Scatterplot of Midwest Area vs Population", fontsize=14)
plt.legend(fontsize=8)    
plt.show()   

# # Alternate Syntax based on plt
# plt.title("Scatterplot of Midwest Area vs Population", fontsize=22)
# plt.xlabel("Area", fontsize = 12)
# plt.ylabel("Population", fontsize = 12)
# plt.xlim(0, 45)
# plt.ylim(0, 90000)
# plt.show()
````

Pair plot is used to plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

It is also possible to show a subset of variables or plot different variables on the rows and columns.

__When to use pair plot?__

Pair plots are used to:
- Do the univariate and bivariate analysis with 1 line of code

Use `sns.pairplot()` to draw the pair plot
