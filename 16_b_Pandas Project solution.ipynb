```python
# Import libraries
import pandas as pd
```

```python
# Read the life expectancy data
df = pd.read_csv('Datasets/Life Expectancy Data.csv')
df
```

### A1. Display the first five rows of 'Life Expectancy Data' and make it show all the columns without truncating.

```python
pd.set_option('display.max_columns', 50)
df.head(5)
```

### A3. Extract the number of rows and columns present in the data

```python
# Get the shape of df
df.shape

print("Number of Rows : ", df.shape[0])
print("Number of Columns : ", df.shape[1])
```

### A4. Does any of columns have extra trailing spaces in the name? If so, remove the extra spaces

```python
# Print the column names of df
df.columns
```

__Inference__
- Few of the columns have extra spaces in the names like BMI, HIV/AIDS etc. 

Let's remove these

```python
# Remove the extra space from column names
df = df.rename(columns=lambda x: x.strip())

df.columns
```

### A5. Get the summary of the data

```python
# Get the summary of df
df.describe()
```

### A6. Are there any missing values in the data? If so, which columns have more than 10% missing values?

```python
# Missing values in the data
df_miss = df.isna().sum().sort_values(ascending=False)
df_miss
```

```python
df_miss = df_miss/df_miss.sum()
df_miss
```

__Inference__
- There are multiple columns having missing values  

Let's get the column which has more than 10% missing

```python
df_miss[df_miss > .1]
```

### A7. Treat the missing values by replacing them with the column median

Replace missing values with the median.

```python
# Replace missing values using median. Except country and status column as they are categorical columns and doesn't have any missing values
for i in df.columns.drop(['Country', 'Status']):
    df[i].fillna(df[i].median(), inplace = True)
```

```python
df.isna().sum()
```

### A8. How many countries have the status as developed and developing respectively

```python
# Get the shape of df
status_df = df[['Country','Status']].drop_duplicates()
status_df.head()
```

```python
status_df['Status'].value_counts()
```

__Inference__
- 161 countries are developing and 32 developed 

### A9. Which country has the highest BMI (average)

```python
# Get the average BMI of all countires
bmi_df = df.groupby('Country').agg({'BMI': 'mean'}).reset_index()

bmi_df.head()
```

```python
# Arrange the dataframe by value of BMI
bmi_df = bmi_df.sort_values(by=['BMI'], ascending=False).reset_index(drop = True)
```

```python
bmi_df
```

```python
bmi_df['Country'][0]
```

__Inference__
- Nauru has the highest BMI

### A10. Who has higher life expectancy, developed or developing countries?

```python
# Get the average life expectancy among developed and developing countries
bmi_df = df.groupby('Status').agg({'Life expectancy': 'mean'}).reset_index()

bmi_df.head()
```

__Inference__
- Developed countries have high life expectancy (79.2) as compared to developing countires (67.1)

### A11. Which year experienced the lowest adult mortality

```python
# Get the average life expectancy among developed and developing countries
adult_mortality_by_year = df.groupby('Year').agg({'Adult Mortality': 'mean'}).reset_index()
adult_mortality_by_year
```

```python
adult_mortality_by_year.loc[adult_mortality_by_year['Adult Mortality'] == adult_mortality_by_year['Adult Mortality'].min(), 'Year']
```

### A12. What is the average life expectancy when the GDP is above 600

```python
# Get the rows where GDP is above 600
high_gdp_df = df[df['GDP'] > 600]

high_gdp_df['Life expectancy'].mean()
```
